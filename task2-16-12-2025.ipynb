```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.linspace(0,5,12)
y=x**2
plt.plot(x,y)
plt.show()
```

```python
plt.xlabel('X Values')
plt.ylabel('Y Values')
plt.title('Graph of $Y = X^2$')
plt.plot(x,y)
plt.show()
```

```python
X1 = np.linspace(-np.pi,np.pi,256)
X2 = np.linspace(-np.pi,np.pi,256)
Y1 = np.sin(X1)
Y2 = np.cos(X2)
plt.plot(X1, X2, Y1, Y2)
plt.show()
```

```python
plt.figure(figsize=(10,6),dpi=70)
plt.plot(X1, Y1)
plt.plot(X2, Y2)
plt.show()
```

```python
plt.figure(figsize=(10,6),dpi=80)
plt.plot(X1, Y1, 'r-.')
plt.plot(X2, Y2, 'b--')
plt.show()
```

```python
plt.figure(figsize=(10,6),dpi=80)
plt.plot(X1, Y1, color='r')
plt.plot(X2, Y2, color='b')
plt.show()
```

```python
plt.figure(figsize=(10,6),dpi=80)
plt.plot(X1, Y1, color='r',lw=2.6)
plt.plot(X2, Y2, color='b',lw=2.6)
plt.show()
```

```python
plt.figure(figsize=(10,6),dpi=80)
plt.plot(X1, Y1, color='r',lw=2.6,ls=':')
plt.plot(X2, Y2, color='g',lw=2.6,ls='--')
plt.show()
```

```python
plt.figure(figsize=(10,6),dpi=80)
plt.plot(X1, Y1, color='r',lw=2.6,ls=':')
plt.plot(X2, Y2, color='g',lw=2.6,ls='--')
plt.xticks([-np.pi,-np.pi/2,0,np.pi/2,np.pi])
plt.yticks([-1,0,1])
plt.show()
```
